```python
features = ['Accelerometer X', 'Accelerometer Y', 'Accelerometer Z', 'Temperature',
       'Gyroscope X', 'Gyroscope Y', 'Gyroscope Z', 'Magnetometer X',
       'Magnetometer Y', 'Magnetometer Z']

X = df[features]

y = df.label
```

```python
from sklearn.pipeline import make_pipeline as MP
from sklearn.preprocessing import Imputer as Imp
from sklearn.neighbors import KNeighborsClassifier as KN

my_pipeline = MP(Imp(), KN(n_neighbors=1))
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(my_pipeline, X, y, scoring='accuracy')
print(scores)
```
